```python
import pandas as pd
import numpy as np
```

```python
# Clean raw dengue data
df_dengue = pd.DataFrame()
for year in range(2009, 2018):
    temp = pd.read_csv(f'../data/raw/2009_dengue.csv')
    # drop NaN rows
    temp = temp[temp.iloc[:,0].notna()]
    # remove district row
    temp = temp.iloc[1:]
    # transpose to use districts as columns
    temp = pd.DataFrame(temp.transpose())
    # use 1st row as column headers
    temp.columns = temp.iloc[0]
    temp.columns.name = None
    # remove 1st row
    temp = temp.iloc[1:-1].reset_index(drop=True)
    index = pd.date_range(start=f'1/1/{year}', periods=52, freq='W')
    temp['index'] = index
    temp = temp.set_index('index')
    df_dengue = pd.concat([df_dengue, temp])
df_dengue
```

```python
df_dengue.to_csv('../data/clean/dengue_cases_by_district.csv')
```
